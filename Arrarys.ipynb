#Arrays Using numpy

```python
import numpy as np

a = np.array([1,2,3,5,7],dtype='i')

b = np.array((2,3,5),dtype='f')

print(a)
print(type(a))
print(a[0:2])

print(b)
print(type(b))
print(b[1:2])
```

```python
import numpy as np
a = np.array([[1,2,3],[4,5,6]])

print(type(a))
print(a)

print(a[0])

print(a[0,2])
print(a[1,1])
```

```python
c = np.array([[[1,2,3],[4,5,6],[0,0,-1]],[[-1,-2,-3],[-4,-5,-6],[0,0,1]]])

print(type(c))
print(c)
```

```python

print(c.shape)
print(c.shape[0])
print(c.shape[1])
print(c.shape[2])

```
