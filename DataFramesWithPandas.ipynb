#Data Frames  using Pandas

```python
import pandas as pd
import numpy as np

print(pd.__version__)
```

```python
A = pd.Series([2,3,4,5],index=['a','b','c','d'])

print(A)

print(A.values)
```

```python
grads_dict = {'A':4,'B':3.5,'C':3,'D':2.5}
grads = pd.Series(grads_dict)

print(grads)

print(grads.values)
```

```python
marks_dict = {'A':85,'B':75,'C':65,'D':55}
marks = pd.Series(marks_dict)

print(marks)

print(marks.keys)

print(marks.values)
```

```python
D = pd.DataFrame({'Marks':marks,'Grades':grads})

print(D)

print("Transpose of D")
print(D.T)

print("Keys")
print(D.keys)

print("Values")
print(D.values)

print("Colums")
print(D.columns)
```

```python
print("Adding new column")
D['ScaledMarks'] = 100*(D['Marks']/90)

print(D)

print(D.columns)
```

```python
print("Print the table")

D
```

```python
print("Searching for Marks > 70")
G = D[D['Marks']>70]

print(G)
```

##Slicing

```python
# Initializing the nested list with Data set
player_list = [['M.S.Dhoni', 36, 75, 5428000], 
               ['A.B.D Villers', 38, 74, 3428000], 
               ['V.Kholi', 31, 70, 8428000],
               ['S.Smith', 34, 80, 4428000], 
               ['C.Gayle', 40, 100, 4528000],
               ['J.Root', 33, 72, 7028000],
               ['K.Peterson', 42, 85, 2528000]]
  
# creating a pandas dataframe
df = pd.DataFrame(player_list, columns=['Name', 'Age', 'Weight', 'Salary'])
  
# data frame before slicing
df
```

```python
df1 = df.iloc[:,:]

df1
```

```python
# Slicing rows in data frame
df1 = df.iloc[1:4]
  
# data frame after slicing
df1
```

```python
# Slicing columns in data frame
df1 = df.iloc[:,1:3]
  
# data frame after slicing
df1
```

```python
# Slicing rows and columns in data frame
df1 = df.iloc[1:4,1:3]

df1
```

```python
# Specific row
print("3rd row")
df.iloc[2,:]
```

```python
# Specific column
print("3rd colum")
df.iloc[:,2]
```

```python
print("Reversing the table")
df.iloc[::-1,:]
```

```python
print("Last column values")
df.iloc[::,-1]
```

##Filling Missing values

```python
# dictionary of lists
dict = {'First Score':[100, 90, 85, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score':[56, 40, 80, 98],
        'Fourth Score': [23,45,np.nan,76],
        'Fifth Score': [54,76,92,np.nan],
        'Sixth Score': [np.nan,53,67,59]}
  
# creating a dataframe from dictionary
df = pd.DataFrame(dict)

df
```

```python
# using isnull() function  
df.isnull()
```

```python
# using notnull() function 
df.notnull()
```

```python
print("Filling missing values using column")
df['Second Score'] = df['Second Score'].fillna(0)

df
```

```python
df['Fourth Score'] = df['Fourth Score'].replace(np.nan, 0)

df
```

```python
# using notnull() function 
df.notnull()
```

```python
print("Filling missing values in the entire table using fillna or replace")
df = df.replace(np.nan,0)

df
```

```python
print("Filling missing values in the entire table using fillna or replace")
df = df.fillna(0)

#df.replace(np.nan,0)
df
```

##Dropping Missing Values

```python
# dictionary of lists
dict = {'First Score':[100, 90, 85, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score':[56, 40, 80, 98],
        'Fourth Score': [23,45,76,np.nan],
        'Fifth Score': [54,76,92,np.nan]}
  
# creating a dataframe from dictionary
df1 = pd.DataFrame(dict)

df1
```

```python
# using dropna() function withput specificing axis  
df1.dropna()

df1
```

```python
# using dropna() function to drop rows having na      
df1.dropna(axis = 0)
```

```python
# making new data frame with dropped NA values 
new_df1 = df1.dropna(axis = 0, how ='any') 
    
new_df1
```

```python
# using dropna() function to drop columns having na      
df1.dropna(axis = 1)
```

```python
# using dropna() function to drop columns having na      
new_dfc = df1.dropna(axis = 1)

new_dfc
```
