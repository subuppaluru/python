##Conditional selection (boolean arrays)

```python
g7_pop
```

```python
g7_pop > 70
```

```python
g7_pop[g7_pop > 70]
```

```python
g7_pop[(g7_pop > 80) | (g7_pop < 40)]
```

```python
g7_pop[(g7_pop > 80) & (g7_pop < 200)]
```

```python
g7_pop.mean()
```

```python
g7_pop[g7_pop > g7_pop.mean()]
```

```python
g7_pop.std()
```

```python
g7_pop[(g7_pop > g7_pop.mean() - g7_pop.std() / 2) | (g7_pop > g7_pop.mean() + g7_pop.std() / 2)]
```
